```python
import numpy as np
import pandas as pd

def Generate_Random_Data(n_samples=200, n_features=50):
    try:
        # Generate random feature matrix
        X = np.random.randn(n_samples, n_features)

        # Generate true coefficients (only positive values)
        true_coef = 3 * np.random.randn(n_features)
        true_coef[true_coef < 0] = 0  # Set negative coefficients to 0

        # Generate target variable y
        y = np.dot(X, true_coef)  # Linear combination
        y += 5 * np.random.normal(size=(n_samples,))  # Add noise

        # Convert X into a DataFrame with column names
        feature_columns = [f"Feature_{i+1}" for i in range(n_features)]
        X_df = pd.DataFrame(X, columns=feature_columns)

        # Convert y into a DataFrame
        y_df = pd.DataFrame(y, columns=["Target"])

        # Combine feature matrix (X) and target (y) into one table
        result_df = pd.concat([X_df, y_df], axis=1)

        return result_df  # âœ… Now returns a table instead of raw arrays

    except Exception as e:
        print(f"Error: {e}")
        return None

# Run the function
random_data_table = Generate_Random_Data(n_samples=200, n_features=50)

# Print the first 5 rows to check the table
random_data_table.head()
```
