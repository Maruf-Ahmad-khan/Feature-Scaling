```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\InhouseInternship\majorproject7\Test set.csv")
```

```python
def Information(df: pd.DataFrame)->pd.DataFrame:
     
     """
     Show the 
     data Distribution
     and correlation
     
     """
     
     try:
          return df.describe()
     
     except Exception as e:
          
          print(f"Error: {e}")
          
          
def Description(df: pd.DataFrame)->pd.DataFrame:
     
     """
     Use skimpy
     
     """
     from skimpy import skim
     try:
          return skim(df)
     
     except Exception as e:
          print(f"Error: {e}")
          

def Summary_Tools(df: pd.DataFrame)->pd.DataFrame:
     
     """
     Use Summary Tools
     
     """
     try:
          from summarytools import dfSummary
          return dfSummary(df)
     
     except Exception as e:
          print(f"Error: {e}")
          
          

Information(df)
     
```

```python
def Pair_Plot(df: pd.DataFrame)->None:
     
     """
     Plot the
     Pair Plot
     
     """
     
     try:
          import seaborn as sns
          sns.set(font_scale= 2)
          sns.pairplot(data = df,
                       y_vars= ['Weight'],
                       x_vars= ['Height'],
                       height= 6
                       )
          
          plt.show()
          
     except FileNotFoundError:
          raise FileNotFoundError(f"Errors")
     
     except Exception as e:
          print(f"Errors: {e}")
          
Pair_Plot(df)
```

```python
def Select_cols(df: pd.DataFrame)->pd.DataFrame:
     
     """
     Select 
     seperate
     dataframe
     
     """
     try:
          return df.filter(items=['Weight', 'Height'])
     
     except Exception as e:
          print(f"Error: {e}")
          
data_cols = Select_cols(df)
data_cols.head()
```

```python
# Correlation matrix

def Correlation_Matrix(data_cols: pd.DataFrame)->None:
     
     """
     Plot the
     Heat map
     
     """
     try:
          import seaborn as sns
          sns.set(font_scale=2)
          sns.heatmap(
                      data_cols.corr(),
                      annot = True,
                      fmt = '.2g',
                      center = 0,
                      cmap = 'coolwarm',
                      linewidths = 1,
                      linecolor = 'black'
                      )
          plt.show()
          
     except Exception as e:
          
          print(f"Error: {e}")
          
Correlation_Matrix(data_cols)
     
```

```python
# Isolate X And Y

def Isolate_X(data_cols: pd.DataFrame):
     
     """
     Isolate
     X And Y
     
     """
     try:
          X = data_cols.Weight 
          return X
     
     except Exception as e:
          print(f"Error: {e}")
          
          
def Isolate_Y(data_cols: pd.DataFrame):
     
     try:
          Y = data_cols.Height
          return Y
     except Exception as e:
          print(f"Error: {e}")

X = Isolate_X(data_cols)
X
```

```python
Y = Isolate_Y(data_cols)
Y
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np
# Adding Constant

def Adding_constant(X):
     
     """_summary_
        Add the 
        Constant.
        and this
        constant 
        is always 1.
     """
     
     try:
          X = sm.add_constant(X)
          return X
     except Exception as e:
          print(f"Error: {e}")
          
x = Adding_constant(X)
x.head()
```

```python
# Regression model

def Model_1(X, Y):
     
     """_summary_
        Regression
        Model.
     """
     try:
          model_1 = sm.OLS(endog = Y, exog = X).fit()
          return model_1.summary()
     
     except Exception as e:
          print(f"Error: {e}")

print("The OLS Model is given below: \n")        
print(Model_1(X, Y))
```

```python
# plot the graph

def Visualization(X, Y):
     
     """_summary_
        plot the graph.
     """
     
     try:
          X = data_cols.Weight
          plt.plot(X, Y, 'o')
          b, a = np.polyfit(X, Y, 1)
          plt.plot(X, b * X + a)
          plt.title("Weight vs Height", fontsize = 13, weight = 'bold')
          plt.xlabel('Carat', fontsize = 13, weight = 'bold')
          plt.ylabel('Price', fontsize = 13, weight = 'bold')
          plt.show()
          
     except Exception as e:
          print(f"Error: {e}")
          
Visualization(X, Y)
```
