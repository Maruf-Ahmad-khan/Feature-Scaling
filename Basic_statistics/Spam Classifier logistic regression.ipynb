```python
# Summary statistics

def Information(df: pd.DataFrame)->pd.DataFrame:
     
     """
     Show the 
     data Distribution
     and correlation
     
     """
     
     try:
          return df.describe()
     
     except Exception as e:
          
          print(f"Error: {e}")
          
          
def Description(df: pd.DataFrame)->pd.DataFrame:
     
     """
     Use skimpy
     
     """
     from skimpy import skim
     try:
          return skim(df)
     
     except Exception as e:
          print(f"Error: {e}")
          

def Summary_Tools(df: pd.DataFrame)->pd.DataFrame:
     
     """
     Use Summary Tools
     
     """
     try:
          from summarytools import dfSummary
          return dfSummary(df)
     
     except Exception as e:
          print(f"Error: {e}")
          
          

Information(df)
     
```

```python
def Histogram(df: pd.DataFrame)->pd.DataFrame:
     
     try:
          
          import matplotlib.pyplot as plt
          import seaborn as sns
          df.hist(figsize=(30,20))
          plt.show()
          
     except Exception as e:
          print(f"Error: {e}")
          return None
     
Histogram(df)
```

```python
def Correlations(df: pd.DataFrame)->pd.DataFrame:
     
     try:
          
          import seaborn as sns
          import matplotlib.pyplot as plt
          plt.figure(figsize=(30,20))
          sns.heatmap(df.corr(),
                      annot=True,
                      fmt='.1g',
                      center= 0,
                      cmap='coolwarm', 
                      linecolor='black',
                      linewidths=1
                      )
          plt.show()
          
     except Exception as e:
          print(f"Error: {e}")
          return None
     
Corrs = Correlations(df)
Corrs
```

```python
def Y_Var(df:pd.Series)->pd.Series:
     
     try:
          
          return df.spam
     except Exception as e:
          print(f"Error: {e}")
          return None
     
def X_Var(df: pd.DataFrame)->pd.DataFrame:
     
     try:
          return df.iloc[:, :-1]
     except Exception as e:
          print(f"Error: {e}")
          return None
     
def Constant(X:pd.DataFrame)->pd.DataFrame:
      
     try:
          import statsmodels.api as sm
          return sm.add_constant(X) 
      
     except Exception as e:
          print(f"Error: {e}")
          return None
     
     
              
     
y = Y_Var(df)
y     
```

```python
from sklearn.model_selection import train_test_split
def Split_Data_Set(constant:pd.DataFrame, y:pd.Series)->pd.DataFrame:
     
     try:
          X_train, X_test, y_train,  y_test = train_test_split(constant,
                                                               y, 
                                                               test_size=0.2,
                                                               random_state=1502
                                                            )
          
          return X_train, X_test, y_train, y_test
     
          
     except Exception as e:
          print(f"Error: {e}")
          return None, None, None, None
     
X_train, X_test, y_train,  y_test =  Split_Data_Set(constant, y)
X_train
     
```

```python
import warnings
warnings.filterwarnings('ignore')
def Logistic_Regression(y_train:pd.Series, X_train:pd.DataFrame)->pd.DataFrame:
     
     try:
          import statsmodels.api as sm
          model = sm.Logit(y_train, X_train).fit()
          return model
     except Exception as e:
          print(f"Error: {e}")
          return None, None
     

  
model = Logistic_Regression(y_train, X_train)
print(model.summary())        
     
```

```python
def logistic_reader(coefficient: float)->float:
     
     try:
          
     
          import numpy as np
          probability = round((np.exp(coefficient) - 1) * 100,2)
          
          if probability > 0:
               print("The likelihood increase by", probability, '%')
               
          elif probability == 0:
               print("No Impact")
               
          else:
               print("The likelihood decreases by", probability, "%")
               
     except Exception as e:
          print(f"Error: {e}")
          return None
     
logistic_reader(2.7849)  
```

```python
from sklearn.metrics import confusion_matrix
import numpy as np
def Confusion_Matrix(y_test:pd.DataFrame, prediction:np.ndarray)->np.ndarray:
     
     try:
          print(confusion_matrix(y_test, prediction))
          tn, fp, fn, tp = confusion_matrix(y_test, prediction).ravel()
          return tn, fp, fn, tp
     
     except Exception as e:
          print(f"Error: {e}")
          return None
     
tn, fp, fn, tp = Confusion_Matrix(y_test, prediction)
tn, fp , fn, tp 
```

```python
def Accuracy(tn:np.ndarray,
             fp:np.ndarray,
             fn:np.ndarray,
             tp:np.ndarray
             )->np.ndarray:
     
     try:
          accuracy = (tp + tn) / (tp + tn + fp + fn)
          return accuracy
     
     except Exception as e:
          print(f"Error: {e}")
          return None
     
accuracy = Accuracy(tn, fp, fn, tp)
print(f"Accuracy is : {round(accuracy * 100, 1)} %")
     
```

```python
def F_1_Score(fp:np.ndarray,
             fn:np.ndarray,
             tp:np.ndarray
             )->np.ndarray:
     
     try:
          f1 = 2 * tp / (2 * tp + fn + fp)
          return f1
     
     except Exception as e:
          print(f"Error: {e}")
          return None
     
f1_score = F_1_Score(fp, fn, tp)
print(f"F1 Score is: {round(f1_score * 100, 1)} %")
```

```python
def Specificity(tn: np.ndarray, 
                fp: np.ndarray
                )->np.ndarray:
     
     try:
          
          specificity = tn / (tn + fp)
          return specificity
     
     except Exception as e:
          print(f"Error: {e}")
          return None
     
def Sensitivity(tp: np.ndarray,
                fn: np.ndarray
                )->np.ndarray:
     
     try:
          sensitivity = tp / (tp + fn)
          return sensitivity
     except Exception as e:
          print(f"Error: {e}")
          return None
     
specificity = Specificity(tn, fp)
print(f"specificity is : {round(specificity * 100, 1)} %")
sensitivity = Sensitivity(tp, fn)
print(f"sensitivity is : {round(sensitivity * 100, 1)} %")
```
