```python
# Summary statistics

def Information(data: pd.DataFrame)->pd.DataFrame:
     
     """
     Show the 
     data Distribution
     and correlation
     
     """
     
     try:
          return data.describe()
     
     except Exception as e:
          
          print(f"Error: {e}")
          
          
def Description(data: pd.DataFrame)->pd.DataFrame:
     
     """
     Use skimpy
     
     """
     from skimpy import skim
     try:
          return skim(data)
     
     except Exception as e:
          print(f"Error: {e}")
          

def Summary_Tools(data: pd.DataFrame)->pd.DataFrame:
     
     """
     Use Summary Tools
     
     """
     try:
          from summarytools import dfSummary
          return dfSummary(data)
     
     except Exception as e:
          print(f"Error: {e}")
          
          

Information(data)
     
```

- T-Test

```python
def T_test(data:pd.DataFrame)->pd.DataFrame:
     
     try:
          group_1 = data.where(data.catholic == 0).dropna()['income']
          group_2 = data.where(data.catholic == 1).dropna()['income']
          stat , p = ss.ttest_ind(group_1, group_2)
          return stat, p
     
     except ValueError:
          raise ValueError("Value Error")
     
     except Exception as e:
          print(f"Error: {e}")
          return None
     
     
T_test(data) 
```

- Get the continuous variables

```python
def Continuous_Variable(data: pd.DataFrame)->pd.DataFrame:
     
     try:
          contnuous = ['number_places_lived',
                        'mom_age', 'dad_age',
                        'mom_score', 'dad_score',
                        'income'
                        ]
          
          stat = {}
          p = {}
          for x in contnuous:
               group_1 = data.where(data.catholic == 0).dropna()[x]
               group_2 = data.where(data.catholic == 1).dropna()[x]
               stat[x], p[x] = ss.ttest_ind(group_1, group_2)
               
          ttests = pd.DataFrame.from_dict(p, orient = "index")
          ttests.columns = ["p-value"]
          return ttests
     
     
     except Exception as e:
          print(f"Error: {e}")
          return None
     
Continuous_Variable(data)
     
```

```python
def Chisquare_test(data:pd.DataFrame)->pd.DataFrame:
     
     try:
          
          tab = pd.crosstab(index = data.poverty,
                            columns = data.catholic
                            )
          stat, p, dof, exp = ss.chi2_contingency(tab)
          return stat, p, dof, exp
     
     except Exception as e:
          print(f"Error: {e}")
          return None
     
     
Chisquare_test(data)
print(p, '\n', stat, '\n', dof, '\n',  exp, '\n')
```

```python
def Chi_Test(data:pd.DataFrame)->pd.DataFrame:
     
     try:
          
          categorical = ['poverty', 'food_stamps']
          p = {}
          for y in categorical:
               tab = pd.crosstab(index = data.poverty,
                                 columns = data.catholic
                                 )
               stat, p[y], dof, exp = ss.chi2_contingency(tab)
          chisquare = pd.DataFrame.from_dict(p, orient = 'Index')
          chisquare.columns = ['p_value']
          return chisquare, stat, p[y], dof, exp
     
     except Exception as e:
          print(f"Error: {e}")
          return None
     
Chi_Test(data)
```
