```python
def Dependent_Variable(data: pd.DataFrame)->pd.DataFrame:
     
     try:
          import numpy as np
          data['yesno'] = np.where(data["yesno"] == "y", 1, 0)
          return data
     
     except Exception as e:
          print(f"Error: {e}")
          return None
     
     
dependent = Dependent_Variable(data)
dependent.head(1)
```

- Prepare X and Y

```python
def Y_Var(dependent:pd.Series)->pd.Series:
     
     try:
          
          return dependent.yesno
     except Exception as e:
          print(f"Error: {e}")
          return None
     
def X_Var(dependent: pd.DataFrame)->pd.DataFrame:
     
     try:
          return dependent.iloc[:, :-1]
     except Exception as e:
          print(f"Error: {e}")
          return None
     
def Constant(X:pd.DataFrame)->pd.DataFrame:
      
     try:
          import statsmodels.api as sm
          return sm.add_constant(X) 
      
     except Exception as e:
          print(f"Error: {e}")
          return None
     
     
              
     
y = Y_Var(dependent)
y     
```

```python
X = X_Var(dependent)
X
```

```python
constant = Constant(X)
constant
```

- Training and test set

```python
from sklearn.model_selection import train_test_split
def Split_Data_Set(constant:pd.DataFrame, y:pd.Series)->pd.DataFrame:
     
     try:
          X_train, X_test, y_train,  y_test = train_test_split(constant,
                                                               y, 
                                                               test_size=0.2,
                                                               random_state=1502
                                                            )
          
          return X_train, X_test, y_train, y_test
     
          
     except Exception as e:
          print(f"Error: {e}")
          return None, None, None, None
     
X_train, X_test, y_train,  y_test =  Split_Data_Set(constant, y)
X_train
     
```

- Fit the LR Algo

```python
def Logistic_Regression(y_train:pd.Series, X_train:pd.DataFrame)->pd.DataFrame:
     
     try:
          import statsmodels.api as sm
          model = sm.Logit(y_train, X_train).fit()
          return model
     except Exception as e:
          print(f"Error: {e}")
          return None, None
     

  
model = Logistic_Regression(y_train, X_train)
print(model.summary())        
     
```

- Function to read the Coefficients

```python
def logistic_reader(coefficient: float)->float:
     
     try:
          
     
          import numpy as np
          probability = round((np.exp(coefficient) - 1) * 100,2)
          
          if probability > 0:
               print("The likelihood increase by", probability, '%')
               
          elif probability == 0:
               print("No Impact")
               
          else:
               print("The likelihood decreases by", probability, "%")
               
     except Exception as e:
          print(f"Error: {e}")
          return None
     
logistic_reader(0.0400)  
```

- confusion matrix

```python
from sklearn.metrics import confusion_matrix
import numpy as np
def Confusion_Matrix(y_test:pd.DataFrame, prediction:np.ndarray)->np.ndarray:
     
     try:
          print(confusion_matrix(y_test, prediction))
          tn, fp, fn, tp = confusion_matrix(y_test, prediction).ravel()
          return tn, fp, fn, tp
     
     except Exception as e:
          print(f"Error: {e}")
          return None
     
tn, fp, fn, tp = Confusion_Matrix(y_test, prediction)
tn, fp , fn, tp 
```

- Accuracy

```python
def Accuracy(tn:np.ndarray,
             fp:np.ndarray,
             fn:np.ndarray,
             tp:np.ndarray
             )->np.ndarray:
     
     try:
          accuracy = (tp + tn) / (tp + tn + fp + fn)
          return accuracy
     
     except Exception as e:
          print(f"Error: {e}")
          return None
     
accuracy = Accuracy(tn, fp, fn, tp)
print(f"Accuracy is : {round(accuracy * 100, 1)} %")
     
```

- F1 Score

```python
def F_1_Score(fp:np.ndarray,
             fn:np.ndarray,
             tp:np.ndarray
             )->np.ndarray:
     
     try:
          f1 = 2 * tp / (2 * tp + fn + fp)
          return f1
     
     except Exception as e:
          print(f"Error: {e}")
          return None
     
f1_score = F_1_Score(fp, fn, tp)
print(f"F1 Score is: {round(f1_score * 100, 1)} %")
```

- specificiy and sensitivity

```python
def Specificity(tn: np.ndarray, 
                fp: np.ndarray
                )->np.ndarray:
     
     try:
          
          specificity = tn / (tn + fp)
          return specificity
     
     except Exception as e:
          print(f"Error: {e}")
          return None
     
def Sensitivity(tp: np.ndarray,
                fn: np.ndarray
                )->np.ndarray:
     
     try:
          sensitivity = tp / (tp + fn)
          return sensitivity
     except Exception as e:
          print(f"Error: {e}")
          return None
     
specificity = Specificity(tn, fp)
print(f"specificity is : {round(specificity * 100, 1)} %")
sensitivity = Sensitivity(tp, fn)
print(f"sensitivity is : {round(sensitivity * 100, 1)} %")
```
