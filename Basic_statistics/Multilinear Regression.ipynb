```python
# Summary statistics

def Generate_summary(data: pd.DataFrame)->pd.DataFrame:
     
     """
     Generate summary
     using Skimpy
     
     """
     from skimpy import skim

     try:
          return skim(data)
     
     except Exception as e:
          print(f"Error: {e}")
          
def Summary_tools(data: pd.DataFrame)->pd.DataFrame:
     
     """
     Generate depth
     Summary Using
     Summary_tools
     
     """
     from summarytools import dfSummary
     try:
          return dfSummary(data)
     except Exception as e:
          print(f"Errors : {e}")
          
def Describe(data: pd.DataFrame)->pd.DataFrame:
     
     """
     Use Pandas
     Describe
     methods
     
     """
     try:
          return data.describe()
     
     except Exception as e:
          print(f"Error: {e}")
     
          
Summary = Generate_summary(data)
Summary
```

```python
# plotting continuous IV against dependent variable

def Pair_Plot(data: pd.DataFrame)->None:
     
     """
     Plot the
     Pair Plot
     
     """
     
     try:
          sns.set(font_scale= 2)
          sns.pairplot(data = data,
                       y_vars= ['salary'],
                       x_vars= ['yrs.service',
                                'yrs.since.phd'],
                       height= 6
                       )
          
          plt.show()
          
     except FileNotFoundError:
          raise FileNotFoundError(f"Errors")
     
     except Exception as e:
          print(f"Errors: {e}")
          
Pair_Plot(data)
```

```python
# Correlation matrix

def Correlation_Matrix(data_cols: pd.DataFrame)->None:
     
     """
     Plot the
     Heat map
     
     """
     try:
          
          sns.set(font_scale=2)
          sns.heatmap(
                      data_cols.corr(),
                      annot = True,
                      fmt = '.2g',
                      center = 0,
                      cmap = 'coolwarm',
                      linewidths = 1,
                      linecolor = 'black'
                      )
          plt.show()
          
     except Exception as e:
          
          print(f"Error: {e}")
          
Correlation_Matrix(data_cols)
     
```

```python
# Categorical Variable
def Find_Frequency(data: pd.DataFrame)->pd.DataFrame:
     
     """
     Find the 
     Frequency
     
     """
     return data.select_dtypes(
                              include = 'object'
                               ).value_counts(
                              ).reset_index(name = 'count'
                               ).sort_values(by = 'count', 
                              ascending = False
                              )
                               
def Categorical_Var(data: pd.DataFrame):
     
     """
     Return a 
     Series.
     
     """
     try:
          categorical = list(data.select_dtypes(include='object'))
          for cat in categorical:
               print(data[cat].value_counts())
          
     except Exception as e:
          raise Exception(f"Error: {e}")
                               
Find_Frequency(data)
```

```python
# Transform objects into dummies

import pandas as pd

def Dummy_Var(data: pd.DataFrame) -> pd.DataFrame:
    """
    Create dummy variables and ensure binary representation as 0 and 1.
    """
    try:
        # Create dummy variables
        df = pd.get_dummies(data, drop_first=True)
        
        # Convert boolean columns to integers (0, 1)
        df = df.apply(lambda col: col.astype(int) if col.dtype == 'bool' else col)
        
        return df
    except Exception as e:
        print(f"Error: {e}")
        return None
   
D_data = Dummy_Var(data)
D_data
```

```python
# Isolate X and Y
y = D_data.salary / 1000

X = D_data.drop(columns=['salary', 'yrs.service'])

X.head(1)

 
```

```python
# Add a constant

def Add_Constant(X: pd.DataFrame)->pd.DataFrame:
     
     """
     Add a constant
     
     """
     try:
          import statsmodels.api as sm
          X = sm.add_constant(X)
          return X
     except Exception as e:
          print(f"Error: {e}")
          
x = Add_Constant(X)
x.head()
```

```python
# Training and Test Set
from sklearn.model_selection import train_test_split
def Test_Train(x : pd.DataFrame, y : pd.Series):
     
     """
     Split into 
     Train and Test
     
     """
     
     try:
          X_train, X_test, y_train, y_test = train_test_split(
               x, y, test_size= 0.2, random_state= 1502
               )
          return X_train, X_test, y_train, y_test
     
     except Exception as e:
          raise Exception(f"Error: {e}")
     

X_train, X_test, y_train, y_test = Test_Train(x, y)
X_train.head(2)
```

# Multilinear Regression

```python
# Multilinear regression

def Multilinear_Regression(y_train : pd.Series, X_train : pd.DataFrame):
     
     """
     Create Multilinear
     Regression model.
     
     """
     try:
          import statsmodels.api as sm
          model = sm.OLS(y_train, X_train).fit()
          return model
     
     except NameError:
          raise NameError(f"Error:")
     
     except Exception as e:
          print(f"Error: An Exception Error is occurred: {e}")
          return None
     
     
Model_Summary = Multilinear_Regression(y_train, X_train)
print(Model_Summary.summary())
```

# Accuracy Assessment

```python
# Predict

def Predict(y_train : pd.Series, X_train : pd.DataFrame,X_test: pd.Series):
     
     """
     Predict 
     
     """
     try:
          import statsmodels.api as sm
          model = sm.OLS(y_train, X_train).fit()
          predictions = model.predict(X_test)
          return predictions[:4]
     
     
     except Exception as e:
          
          print(f"Errors : {e}")
          return None
     
     
Predicts = Predict(y_train, X_train, X_test)
print("The prediction is: ")
Predicts
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import statsmodels.api as sm
def Errors_assessment(y_test: pd.Series,
                      y_train : pd.Series,
                      X_train:pd.DataFrame,
                      X_test:pd.Series
                      ):
     
     """
     Compute 
     Mean_squared_error
     Mean_absolute_error
     
     """
     try:
          model = sm.OLS(y_train, X_train).fit()
          predictions = model.predict(X_test)
          MAE = round(mean_absolute_error(y_test, predictions), 2)
          MSE = round(np.sqrt(mean_squared_error(y_test, predictions)), 2) 
          return f"MAE: {MAE}", f"MSE: {MSE}"
          
          
     except NameError:
          print("NameError")
          
     except Exception as e:
          print(f"Error: {e}")
          return None
     
Ans = Errors_assessment(y_test, y_train, X_train, X_test)
print("Mean_squared_error And Mean_absolute_error is:\n",Ans)
          
```
