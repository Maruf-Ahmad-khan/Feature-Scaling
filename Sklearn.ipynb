# Binarization

```python
import numpy as np
from sklearn import preprocessing
Input_Array = np.array(
                       [
                         [2.1, -1.9, 5.5], 
                         [-1.5, 2.4, 3.5], 
                         [0.5, -7.9, 5.6], 
                         [5.9, 2.3, -5.8]
                        ]
                       )
```

```python
data_binarized = preprocessing.Binarizer(threshold=0.5).transform(Input_Array)
print("Binarized Data : ")
data_binarized
```

```python
from sklearn import preprocessing
ans = np.array([[1,2,3], [4,5,6]])
data_Binary = preprocessing.Binarizer(threshold=1).transform(ans)
data_Binary
```

# Mean Removal


```python
print("Mean of the data is :", Input_Array.mean(axis=0))
print("STD of the data is : ", Input_Array.std(axis = 0))
Data_Scaled = preprocessing.scale(Input_Array)
print("Mean Removal : ", Data_Scaled.mean(axis = 0))
print("STD Removal : ", Data_Scaled.std(axis = 0))
```
